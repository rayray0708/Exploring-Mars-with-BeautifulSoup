```python
# Visit the website
url = 'https://static.bc-edx.com/data/web/mars_facts/temperature.html'
browser.visit(url)
```

### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
#Scrape the webpage
html = browser.html
```

```python
# Create a Beautiful Soup Object
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Extract all rows of data
all_rows = soup.find_all(class_ = 'data-row')
all_rows[0].find_all() #to access certain elements in list, use the find_all() function to convert elements into a list,
#to access the id, use all_rows[0].find_all()[0]
#to access terrestrial_date, use all_rows[0].find_all()[1] ...
```

```python
all_rows[0].find_all()[6].text
```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Hereâ€™s an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# Create an empty list
Mars_temperature_list = []
# Loop through the scraped data to create a list of rows
for row in all_rows:
    id = row.find_all()[0].text
    terrestrial_date = row.find_all()[1].text
    sol = row.find_all()[2].text
    ls = row.find_all()[3].text
    month = row.find_all()[4].text
    min_temp = row.find_all()[5].text
    pressure = row.find_all()[6].text
    
    temp_dict = {
        'id':id,
        'terrestrial_date': terrestrial_date,
        'sol':sol,
        'ls':ls,
        'month':month,
        'min_temp':min_temp,
        'pressure':pressure}
    
    Mars_temperature_list.append(temp_dict)
        
```

```python
Mars_temperature_list
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
Mars_temp_df = pd.DataFrame(Mars_temperature_list, columns = ['id','terrestrial_date','sol','ls','month','min_temp',\
                                                              'pressure'])
```

```python
# Confirm DataFrame was created successfully
Mars_temp_df.head()
```

```python
# 3. What is the average low temperature by month?
avg_low_temp = Mars_temp_df.groupby('month')['min_temp'].mean()
avg_low_temp
```

```python
#ASK QUESTION!!!
avg_low_temp = pd.DataFrame(avg_low_temp, columns = ['month','min_temp'])
avg_low_temp
```

```python
#ASK QUESTION!!!
avg_low_temp.plot(x='month', y='min_temp')
```

```python
# Create a DataFrame from the grouped result (not necessary, but can be useful for labeling)
avg_low_temp = pd.DataFrame({'month': avg_low_temp.index, 'Temperature in Celsius': avg_low_temp.values})
avg_low_temp.head()
```

```python
# Plot the average temperature by month
plt.bar(avg_low_temp['month'],avg_low_temp['Temperature in Celsius'], color = 'green',\
        tick_label = avg_low_temp['month'])
plt.xlabel('Month')
plt.ylabel('Temperature in Celsius')
plt.title('Mars temperatures by month')
plt.xticks(rotation = 90)
```

```python
Mars_temp_df
```

```python
# Identify the coldest and hottest months in Curiosity's location
#coldest_hottest_months = Mars_temp_df.groupby('month')['min_temp'].mean().sort_values()
coldest_hottest_months = Mars_temp_df.groupby('month')['min_temp'].mean()
coldest_hottest_months
```

```python
#This is the sorted series, this works because in this Series, the month column is the index while
#the min_temp column contains values.
#sort_values() function will sort the values in ascending order if not specified otherwise
coldest_hottest_months.sort_values()
```

```python
#Month is treated as an index column, the sort_values() fundtion is sorting based on the values
#(ie., y-axis or min_temp column). So, use Pandas's plot function to plot this series
coldest_hottest_months.sort_values().plot(kind='bar')
```

## What are the coldest and the warmest months on Mars (at the location of Curiosity)?
The 3rd month has the coldest temperature and the 8th month has the warmest temperature.

## Note: Panda's plot() function vs. matplotlib

Matplotlib is ultimately not reading in the pandas dataframe correctly. If you were to hash out the line of code that creates the dataframe and run in matplotlib it should sort the data.

Since pandas plot() function is designed to work with pandas dataframes, it is able to interpret the sorted values in the dataframe and produce the correct result.

```python
# Sort the DataFrame by 'Mean Min Temp' in descending order
coldest_hottest_months_sorted = coldest_hottest_months.sort_values(by='Temperature in Celsius')
coldest_hottest_months_sorted
```

```python
#Plotting using Pandas
coldest_hottest_months_sorted.plot(kind='bar', x = 'month', y='Temperature in Celsius')
# coldest_hottest_months.sort_values('Temperature in Celsius').plot(kind='bar', x = 'month', y='Temperature in Celsius')
plt.ylabel('Temperature in Celsius')
plt.title('Highest to lowest temperature by month')
```
